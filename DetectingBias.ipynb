# CRIME COUNT BASED ON AREA NAME

```python
group_by_area_crime_count=pd.DataFrame(bias_data.groupby('AREA NAME')['Crm Cd'].count())
group_by_area_crime_count.reset_index(inplace=True)
```

```python
group_by_area_crime_count
```

```python
sorted_crime_count=group_by_area_crime_count.sort_values(by='Crm Cd',ascending=False)
sorted_crime_count
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25,10))
plt.xlabel('Area Name', fontsize=12)
plt.ylabel('Crime Count', fontsize=12)
plt.bar(sorted_crime_count['AREA NAME'],sorted_crime_count['Crm Cd'])
```

```python
lapd_demographics=pd.read_excel('/content/drive/MyDrive/Datasets/LAPD Demographics.xlsx')
lapd_demographics
```

```python
lapd_demographics.sort_values(by='LAPD Area',ascending=True,inplace=True)
lapd_demographics
```

```python
lapd_demographics=lapd_demographics.merge(group_by_area_crime_count,left_on='LAPD Area',right_on='AREA NAME')
lapd_demographics
```

```python
lapd_demographics.sort_values(by='Crm Cd',ascending=False,inplace=True)
lapd_demographics
```

```python
lapd_demographics_crime_race=lapd_demographics[['LAPD Area','Total Population','White','Hispanic/Latino','Black/African American','Asian','Other','Crm Cd']]
lapd_demographics_crime_race
```

```python
from pandas.plotting import table
fig,ax=plt.subplots(figsize=(20,10))
ax.axis('off')

tbl=table(ax,lapd_demographics_crime_race,loc='center')
tbl.auto_set_font_size(False)
tbl.set_fontsize(10)
tbl.scale(1,2)
plt.show()

plt.savefig('lapd_demographics_crime_race.png')
plt.close(fig)
```

```python

lapd_demographics_crime_race['APREC'] = lapd_demographics_crime_race['APREC'].str.lower()
lapd_area['APREC'] = lapd_area['APREC'].str.lower()

total_df = lapd_demographics_crime_race.merge(lapd_area, on='APREC')
```

```python
total_df['APREC']=total_df['APREC'].str.upper()
```

```python
total_df
```

```python
total_df['Crime Density']=total_df['Crm Cd']/total_df['AREA']
```

```python
total_df
```

```python
type(total_df['Total Population'])
```

```python
total_df['Total Population']=total_df['Total Population'].str.replace(',','')
total_df['Total Population']=total_df['Total Population'].astype('int')
total_df
```

```python
population_density=total_df['Total Population']/total_df['AREA']
population_density
```

```python
total_df['Population Density']=population_density
total_df
```

```python
total_df['Crime Rate']=(total_df['Crm Cd']/total_df['Total Population'])*100000
total_df
```

```python
total_df.sort_values(by='Crime Rate',ascending=False,inplace=True)
total_df
```

```python
corr_df=total_df[['Total Population','Crm Cd','Crime Density','Crime Rate']]
corr_df_corr=corr_df.corr()
corr_df_corr
```

```python
from pandas.plotting import table
fig,ax=plt.subplots(figsize=(20,20))
ax.axis('off')

tbl=table(ax,corr_df_corr,loc='center')
tbl.auto_set_font_size(False)
tbl.set_fontsize(10)
tbl.scale(1,2)
plt.show()

plt.savefig('correlationbetweenvalues.png')
plt.close(fig)
```
